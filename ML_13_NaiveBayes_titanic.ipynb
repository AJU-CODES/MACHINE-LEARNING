```python
import pandas as aju
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.naive_bayes import GaussianNB
```

```python
df = aju.read_csv('D:\\MACHINE LEARNING\\DATASETS\\ML_13_NaiveBayes_titanic.csv')
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket',	'Cabin','Embarked'],axis= 'columns' ,inplace = True)
df
```

```python
x = df.drop('Survived',axis='columns')
y = df.Survived
```

```python
x.isnull().sum()
```

```python
df.shape
```

```python
dummy = aju.get_dummies(x.Sex).astype(int)
dummy
```

```python
input = aju.concat([x,dummy],axis = 'columns')
input
```

```python
x_act = input.drop(['Sex','male'],axis = 'columns')
x_act
```

```python
x_act['Age'] = x_act.Age.fillna(x_act.Age.mean())
x_act
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_act,y,test_size = 0.2,random_state = 99)
```

```python
model = GaussianNB()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict_proba(x_test)
```
