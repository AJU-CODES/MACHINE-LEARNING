```python
x = df.drop(['Id','Species','Species_num'],axis = 'columns')
y = df['Species_num']
```

```python
digit = load_digits()
dir(digit)
```

```python
dig = aju.DataFrame(digit.data)
dig
```

```python
dig['target'] = digit.target
```

```python
dig
```

```python
x = dig.drop('target',axis = 'columns')
y = dig.target
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 99)
```

```python
rbf_mod = SVC(kernel='rbf')
```

```python
rbf_mod.fit(x_train,y_train)
```

```python
rbf_mod.score(x_test,y_test)
```

```python
lin_mod = SVC(kernel='linear')
```

```python
lin_mod.fit(x_train,y_train)
```

```python
lin_mod.score(x_test,y_test)
```
