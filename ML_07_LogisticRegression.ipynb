```python
import pandas as aju
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df = aju.read_csv("D:\\MACHINE LEARNING\\DATASETS\\ML_07_LOGREG_insurance_data.csv")
df.head()
```

```python
plt.scatter(df.age,df.bought_insurance,marker = '+',color = 'black')
```

```python
df.shape
```

```python
y = df.bought_insurance
```

```python
x_train,x_test,y_train,y_test = train_test_split(df[['age']],y,test_size = 0.1,random_state = 88)
```

```python
x_train
```

```python
x_test
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.predict([[22]])
```

```python
model.score(x_test,y_test)
```

```python
model.predict_proba(x_test)
```
