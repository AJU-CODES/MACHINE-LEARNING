```python
import pandas as aju
from sklearn.linear_model import LinearRegression
```

```python
df = aju.read_csv("D:\\MACHINE LEARNING\\DATASETS\\ML_05_OneHotEncoder_Homeprices.csv")
df.head()
```

```python
dummy = aju.get_dummies(df.town).astype(int)
dummy
```

```python
file = aju.concat([df,dummy],axis = 'columns')
file
```

```python
final = file.drop(['town','west windsor'],axis = 'columns')
final
```

```python
model = LinearRegression()
model
```

```python
x = final.drop('price',axis = 'columns')
x
```

```python
y = final['price']
y
```

```python
model.fit(x,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3300,0,0]])
```

```python
model.score(x,y)
```
